```python
# corr_pairs = corr_matrix.unstack()
# corr_df = corr_pairs.reset_index()
# corr_df.columns = ['Feature 1', 'Feature 2', 'Correlation']
# corr_df = corr_df[corr_df['Feature 1'] != corr_df['Feature 2']]
# corr_df['Pairs'] = list(zip(corr_df['Feature 1'], corr_df['Feature 2']))
# corr_df['Sorted Pairs'] = corr_df['Pairs'].apply(lambda x: tuple(sorted(x)))
# corr_df = corr_df.drop(columns='Pairs')
# corr_df = corr_df.drop_duplicates(subset='Sorted Pairs').drop(columns='Sorted Pairs')
# corr_df = corr_df.sort_values(by='Correlation', ascending=False)
# print(corr_df)
corr_df = (
    corr_matrix.unstack().reset_index()
    .rename(columns={0: 'Correlation', 'level_0': 'Feature 1', 'level_1': 'Feature 2'})
)

corr_df = corr_df[corr_df['Feature 1'] != corr_df['Feature 2']]
corr_df['Sorted Pairs'] = corr_df.apply(lambda x: tuple(sorted([x['Feature 1'], x['Feature 2']])), axis=1)
corr_df = corr_df.drop_duplicates(subset='Sorted Pairs')
corr_df = corr_df.drop(columns='Sorted Pairs')
corr_df = corr_df.sort_values(by='Correlation', ascending=False)
corr_df.style.hide(axis='index')
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_result = pca.fit_transform(corr_matrix)

plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1])

# Annotate each point
for i, feature in enumerate(corr_matrix.columns):
    plt.text(pca_result[i, 0], pca_result[i, 1], feature)

plt.title('PCA of Correlation Matrix')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid(True)
plt.show()
```

# EDA Insights :
- Bank Deposits, Checking Accounts, and Saving Accounts are highly correlated, indicating they represent similar customer behaviors focused on daily transactions and short-term savings. Customers who use one of these accounts are very likely to use the others.

- Business Lending shows moderate correlation with both Bank Loans and Bank Deposits. This suggests business borrowers often also manage loans and maintain significant deposits, likely due to cash flow and credit management practices.

- Credit Card Balance correlates moderately with Bank Loans and Business Lending. This points to a cluster of products used by customers engaging in credit or borrowing activities, potentially indicating shared financial behavior or risk profiles.

- Estimated Income and Superannuation Savings have low correlations with most other financial features. These two variables appear behaviorally independent, implying that income levels and long-term savings are not strongly tied to how customers use day-to-day banking products.

- PCA confirms these clusters: transactional products (Bank Deposits, Checking, Saving) are grouped tightly, lending and credit features form a separate cluster, and income/superannuation are positioned as outliers.

- The PCA also shows that even among highly correlated features like Saving and Checking Accounts, small differences in how they relate to other variables (like lending or income) can shift their positioning in the PCA plot, explaining minor separations.
